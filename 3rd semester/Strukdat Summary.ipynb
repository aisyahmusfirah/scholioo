```python
# AVL Tree
# Membuat Class Node
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1

# Membuat Class AVLTree
class AVLTree:
    # Fungsi untuk Menyisipkan Node
    def insert(self, root, key):
        # 1. Lakukan penyisipan BST
        if not root:
            return Node(key)
        elif key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        # 2. Update tinggi node ini
        root.height = 1 + max(self.get_height(root.left),
                              self.get_height(root.right))

        # 3. Dapatkan faktor keseimbangan node ini
        balance = self.get_balance(root)

        # 4. Jika node tidak seimbang, maka lakukan rotasi
        # Kasus kiri-kiri
        if balance > 1 and key < root.left.key:
            print("Rotasi kanan pada node:", root.key)
            return self.right_rotate(root)

        # Kasus kanan-kanan
        if balance < -1 and key > root.right.key:
            print("Rotasi kiri pada node:", root.key)
            return self.left_rotate(root)

        # Kasus kiri-kanan
        if balance > 1 and key > root.left.key:
            root.left = self.left_rotate(root.left)
            print("Rotasi kanan pada node:", root.key)
            return self.right_rotate(root)

        # Kasus kanan-kiri
        if balance < -1 and key < root.right.key:
            root.right = self.right_rotate(root.right)
            print("Rotasi kiri pada node:", root.key)
            return self.left_rotate(root)

        return root

    # Fungsi untuk rotasi kanan
    def right_rotate(self, z):
        y = z.left
        T3 = y.right

        y.right = z
        z.left = T3

        z.height = 1 + max(self.get_height(z.left),
                           self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left),
                           self.get_height(y.right))

        return y

    # Fungsi untuk rotasi kiri
    def left_rotate(self, z):
        y = z.right
        T2 = y.left

        y.left = z
        z.right = T2

        z.height = 1 + max(self.get_height(z.left),
                           self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left),
                           self.get_height(y.right))

        return y

    # Mendapatkan tinggi node
    def get_height(self, node):
        if not node:
            return 0
        return node.height

    # Mendapatkan faktor keseimbangan node
    def get_balance(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)

    # Pre-order traversal dari AVL Tree
    def pre_order(self, root):
        if not root:
            return
        print(f"{root.key} ", end="")
        self.pre_order(root.left)
        self.pre_order(root.right)

    # Fungsi untuk menampilkan tree
    def print_tree(self, root, level=0, prefix="Root: "):
        if not root:
            return
        print(" " * (level * 4) + prefix + str(root.key))
        if root.left: self.print_tree(root.left, level + 1, "L--- ")
        if root.right: self.print_tree(root.right, level + 1, "R--- ")

# Contoh Penggunaan
if __name__ == "__main__":
    avl_tree = AVLTree()
    root = None

    # Menyisipkan nilai ke dalam AVL tree
    keys = [10, 85, 15, 70, 20, 60, 30, 50, 65, 80, 90, 40, 5, 55]
    for key in keys:
        print(f"\nMenyisipkan {key} ke dalam AVL Tree:")
        root = avl_tree.insert(root, key)
        avl_tree.print_tree(root)  # Menampilkan pohon sebelum penyeimbangan

    # Menampilkan hasil pre-order traversal
    print("\nPre-order traversal dari AVL tree yang sudah seimbang:")
    avl_tree.pre_order(root)
```
